```python
import pdfplumber

# Path to your PDF
pdf_path = r"C:\Users\apagano\Downloads\eo-info.pdf"

# Open the PDF
with pdfplumber.open(pdf_path) as pdf:
    # Loop through each page
    for page_number, page in enumerate(pdf.pages, start=1):
        # Extract text from the page
        text = page.extract_text()
        if text:  # Check if page has text
            # Split text into lines
            lines = text.split('\n')
            print(f"--- Page {page_number} --------------------------------------------------------------------------------------------------")
            for line in lines:
                if not line.isupper() and not line[0].isdigit() and line != 'Code Description':
                    print(line)
```

```python

# Example: df is your DataFrame : https://pandas.pydata.org/docs/reference/series.html
for col in df.columns:
    col_data = df[col]
    
    print(f"Column: {col}")
    print(f"Data type: {col_data.dtype}")
    print(f"Shape: {col_data.shape}")
    print(f"Missing values: {col_data.isnull().sum()}")
    print(f"Unique values: {col_data.nunique()}")
    print(f"Non-null count (n): {col_data.notnull().sum()}")
    print(f"Null count: {col_data.isnull().sum()}")
    
    # Show range for numeric columns
    if pd.api.types.is_numeric_dtype(col_data):
        print(f"Min: {col_data.min()}")
        print(f"Max: {col_data.max()}")
    else:
        # For strings or categorical, show first 5 unique values
        print(f"Example values: {col_data.dropna().unique()[:5]}")
    
    print("---------------------------------------------------------------------------------------------------------------------------")
```

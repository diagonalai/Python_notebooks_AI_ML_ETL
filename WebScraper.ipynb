```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# List of URLs to scrape
urls = [
    "https://registry.impactgenome.com/"
    # Add more URLs here
]

# Data storage
data = []

# Loop through each URL
for url in urls:
    try:
        response = requests.get(url)
        response.raise_for_status()  # Raise error for bad status codes
        soup = BeautifulSoup(response.text, 'html.parser')

        # Example: Extract the page title and first paragraph
        title = soup.title.string if soup.title else "No title"
        paragraph = soup.find('p').get_text(strip=True) if soup.find('p') else "No paragraph"

        data.append({
            "URL": url,
            "Title": title,
            "First Paragraph": paragraph
        })

    except Exception as e:
        data.append({
            "URL": url,
            "Title": "Error",
            "First Paragraph": str(e)
        })

# Convert to DataFrame
df = pd.DataFrame(data)
df.head()
```

```python
from bs4 import BeautifulSoup
from selenium import webdriver

# Open Chrome and get page
driver = webdriver.Chrome()
driver.get("https://registry.impactgenome.com")

# Get page HTML
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

# Function to recursively print all elements
def print_elements(elem, indent=0):
    if elem.name is not None:  # Only consider tags, skip strings for now
        print('  ' * indent + f"<{elem.name}>")
        for child in elem.children:
            print_elements(child, indent + 1)
    elif str(elem).strip():  # print non-empty strings
        print('  ' * indent + str(elem).strip())

# Start from body
print_elements(soup.body)

driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time

# Paths
chrome_path = r"C:\Users\apagano\Downloads\chrome-win64\chrome-win64\chrome.exe"
chromedriver_path = r"C:\Users\apagano\Downloads\chromedriver-win64\chromedriver.exe"

# Chrome options
options = Options()
options.binary_location = chrome_path
# options.add_argument("--headless")  # optional

# Setup WebDriver
service = Service(chromedriver_path)
driver = webdriver.Chrome(service=service, options=options)

driver.get("https://registry.impactgenome.com/")

# Scroll down to load content
for _ in range(10):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)

# Wait for program cards
programs = WebDriverWait(driver, 15).until(
    EC.presence_of_all_elements_located((By.CSS_SELECTOR, "div.impact-program-card"))
)

data = []
actions = ActionChains(driver)

for program in programs:
    try:
        # Hover to reveal hidden info
        actions.move_to_element(program).perform()
        time.sleep(1)  # small pause to allow content to appear

        # Extract main info
        name = program.find_element(By.TAG_NAME, "h3").text

        # Info icon description (hovered text)
        info_icon = program.find_element(By.CSS_SELECTOR, "i")  # adjust selector if needed
        description = info_icon.get_attribute("title") if info_icon else ""

        # CPO value (adjust selector based on actual HTML)
        cpo_elem = program.find_element(By.CSS_SELECTOR, "span.cpo-value")  # example
        cpo_value = cpo_elem.text if cpo_elem else ""

        data.append({
            "Name": name,
            "Description": description,
            "CPO": cpo_value
        })
    except Exception as e:
        print(f"Error with program: {e}")

# Convert to DataFrame
df = pd.DataFrame(data)
print(df)

driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time

chrome_path = r"C:\Users\apagano\Downloads\chrome-win64\chrome-win64\chrome.exe"
chromedriver_path = r"C:\Users\apagano\Downloads\chromedriver-win64\chromedriver-win64\chromedriver.exe"

options = Options()
options.binary_location = chrome_path
options.add_argument("--start-maximized")  # make window big to load content

service = Service(chromedriver_path)
driver = webdriver.Chrome(service=service, options=options)

driver.get("https://registry.impactgenome.com/")

# Scroll multiple times to load dynamic content
for _ in range(10):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)

# Grab all elements with text
all_elements = driver.find_elements(By.XPATH, "//*[normalize-space(text())]")

data = []
for elem in all_elements:
    text = elem.text.strip()
    if text:  # ignore empty strings
        tag = elem.tag_name
        data.append({'Tag': tag, 'Text': text})

df = pd.DataFrame(data)
print(df.head(50))  # show first 50 rows

driver.quit()
```

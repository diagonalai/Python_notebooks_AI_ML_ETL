```python
#CENSUS API CODEBOOK


import os
import requests
import pandas as pd

# ğŸ“¦ Variables metadata URL
VARIABLES_URL = "https://api.census.gov/data/2023/acs/acs5/variables.json"

# ğŸ›  Download metadata
response = requests.get(VARIABLES_URL)

if response.status_code == 200:
    variables_json = response.json()
    variables_dict = variables_json['variables']

    # Extract key info from each variable
    records = []
    for var_name, var_info in variables_dict.items():
        label = var_info.get('label', '')
        concept = var_info.get('concept', '')
        predicateType = var_info.get('predicateType', '')
        records.append([var_name, label, concept, predicateType])

    # Save as CSV
    df_vars = pd.DataFrame(records, columns=['Variable', 'Label', 'Concept', 'PredicateType'])
    downloads_path = os.path.join(os.path.expanduser("~"), "Downloads")
    output_file = os.path.join(downloads_path, "acs5_variables_2023.csv")
    df_vars.to_csv(output_file, index=False)
    print(f"âœ… All variables metadata saved to: {output_file}")
    print(df_vars.head())

else:
    print("âŒ Error:", response.status_code, response.text)
```

```python
#Census API batch
import os
import requests
import pandas as pd
import time

# ğŸ”‘ API Key
API_KEY = "ba40687ecc4ba036bb4b20af1ebd71999f0b9e60"

# ğŸ“¦ Endpoints
BASE_URL = "https://api.census.gov/data/2023/acs/acs5"
VARIABLES_URL = "https://api.census.gov/data/2023/acs/acs5/variables.json"

# ğŸ“ Save directory
downloads_path = os.path.join(os.path.expanduser("~"), "Downloads")

# ğŸ“¥ STEP 1: Get Full Variable List
response = requests.get(VARIABLES_URL)
variables_json = response.json()
variables_dict = variables_json['variables']

# Skip geography fields and non-estimate fields
all_vars = [var for var in variables_dict.keys() if var.endswith('E') and var not in ('NAME', 'state', 'county')]

print(f"âœ… Total variables found: {len(all_vars)}")

# ğŸ›‘ Set batch size (API request limits)
BATCH_SIZE = 45  # Safe batch size (you can push higher, but ~45 is stable)

# ğŸ“Š STEP 2: Download Data in Batches
batch_number = 1
for i in range(0, len(all_vars), BATCH_SIZE):
    batch_vars = all_vars[i:i + BATCH_SIZE]
    get_vars = ",".join(["NAME"] + batch_vars)

    params = {
        "get": get_vars,
        "for": "county:*",
        "in": "state:*",
        "key": API_KEY
    }

    print(f"ğŸ“¡ Downloading batch {batch_number} ({len(batch_vars)} variables)...")

    try:
        response = requests.get(BASE_URL, params=params)
        if response.status_code == 200:
            data = response.json()
            df = pd.DataFrame(data[1:], columns=data[0])
            df["fips"] = df["state"] + df["county"]

            filename = f"acs5_2023_batch_{batch_number}.csv"
            output_file = os.path.join(downloads_path, filename)
            df.to_csv(output_file, index=False)
            print(f"âœ… Batch {batch_number} saved: {filename}")

        else:
            print(f"âŒ Error batch {batch_number}: {response.status_code}")
            print(response.text)

    except Exception as e:
        print(f"âŒ Failed batch {batch_number}: {str(e)}")

    batch_number += 1

    # ğŸ“‰ Pause to avoid API throttling (optional)
    time.sleep(1)
```

```python
import os
import pandas as pd

import gzip

def compress_csv(input_csv_path, output_gz_path):
    
    with open(input_csv_path, 'rb') as f_in, gzip.open(output_gz_path, 'wb') as f_out:
        f_out.writelines(f_in)

    print(f"Compressed file saved to: {output_gz_path}")

csv_files = [
    r"C:\Users\apagano\Downloads\to_Snowflake\combined_columns.csv",
    r"C:\Users\apagano\Downloads\to_Snowflake\combined_columns_B.csv"
]
    

# Loop over the list to compress each CSV
for csv_file in csv_files:
    output_file = csv_file + '.gz'  # output path, e.g., "file.csv.gz"
    compress_csv(csv_file, output_file)
```

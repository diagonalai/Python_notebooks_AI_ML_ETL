```python
import re

# Step 1: Read file
with open(r"...Desktop\tsc.txt", "r", encoding="utf-8") as file:
    text = file.read()

# Step 2: Replace speaker names with MOD:/RES:
updated_text = text.replace("Haley Cooper", "MOD:").replace("Lilly Fagatele", "RES:")

# Step 3: Remove timestamps (e.g., 0 minutes 3 seconds or 0:03)
updated_text = re.sub(
    r"(?i)\b\d{1,2}\s*minute(?:s)?\s*\d{1,2}\s*second(?:s)?(?:\s*\d{1,2}:\d{2})?\b",
    "",
    updated_text
)
updated_text = re.sub(r"\b\d{1,2}:\d{2}\b", "", updated_text)

# Step 4: Remove standalone HC or LF lines
updated_text = re.sub(r"(?m)^[ \t]*(HC|LF)[ \t]*\r?\n?", "", updated_text)

# Step 5: Merge multiple MOD:/RES: in a row into one
updated_text = re.sub(r"\b(MOD:|RES:)\s+\1\b", r"\1", updated_text, flags=re.IGNORECASE)

# Step 6: Keep MOD:/RES: on the same line as the following text
updated_text = re.sub(r"(?i)(MOD:|RES:)\s*\n\s*", r"\1 ", updated_text)

# Step 7: Remove filler words
for word in ["Mhm.", "Hmm.", "Um.", "mm-hmm.", "Mm-hmm."]:
    updated_text = re.sub(r"\b{}\b".format(re.escape(word)), "", updated_text, flags=re.IGNORECASE)

# Step 8: Clean extra spaces and blank lines
updated_text = re.sub(r"[ \t]{2,}", " ", updated_text)
updated_text = re.sub(r"\n{2,}", "\n\n", updated_text)

# Step 9: Save updated file
with open(r"...example_updated.txt", "w", encoding="utf-8") as file:
    file.write(updated_text)

print("File updated successfully.")
```

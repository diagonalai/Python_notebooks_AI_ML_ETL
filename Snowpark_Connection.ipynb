```python
pip install snowflake-connector-python

import pandas as pd
import snowflake.connector
from snowflake.connector.pandas_tools import write_pandas

# Step 1: Load CSV into a DataFrame
csv_path = "...MoonTea.csv"  
df = pd.read_csv(csv_path)

# Step 2: Define Snowflake connection parameters
conn_params = {
    "user": "YOUR_USERNAME",
    "password": "YOUR_PASSWORD",
    "account": "YOUR_ACCOUNT_ID",  
    "warehouse": "YOUR_WAREHOUSE",
    "database": "YOUR_DATABASE",
    "schema": "YOUR_SCHEMA",
    "role": "YOUR_ROLE"
}

# Step 3: Connect and upload
conn = snowflake.connector.connect(**conn_params)
success, nchunks, nrows, _ = write_pandas(conn, df, "YOUR_TABLE_NAME")

print(f"âœ… Uploaded {nrows} rows to Snowflake table.")
conn.close()

```
